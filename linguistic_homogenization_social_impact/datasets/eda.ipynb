```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
essays_df = pd.read_csv("essays/essays_anon_full.csv")
wassa_df = pd.read_csv("wassa/clean_wassa.csv")
facebook_df = pd.read_csv("facebook/full_dataset_clean.csv")
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

for dataset_name, dataset, ax in zip(
    ["essays", "wassa", "facebook"], [essays_df, wassa_df, facebook_df], axs
):
    texts = dataset["text"]
    print("Dataset: ", dataset_name)
    print("Number of mean tokens: ", np.mean([len(text.split()) for text in texts]))
    print("Number of maximum tokens: ", np.max([len(text.split()) for text in texts]))
    ax.hist([len(text.split()) for text in texts], bins=50)
    print("--" * 20)
```
